```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cor_ours = [0.67, 0.746, 0.858, 0.882, 0.895, 0.914]
cov_ours = [0.887, 0.922, 0.96, 0.958, 0.96, 0.965]
cor_dbm = [0.163, 0.27, 0.569, 0.777, 0.873, 0.952]
cov_dbm = [0.411, 0.626, 0.845, 0.916, 0.947, 0.956]
k = [50, 100, 250, 500, 1000, 15000]

l_k = np.log10(np.array(k))
```

```python
plt.plot(l_k, cor_ours, c = "C0", label = "Correctness - TGT")
plt.plot(l_k, cor_dbm, c = "orange", label = "Correctness - DBM")
plt.plot(l_k, cov_ours, ls = "--", c= "C0", label = "Coverage - TGT")
plt.plot(l_k, cov_dbm, ls = "--", color = "orange", label = "Coverage - DBM")

plt.ylabel("Metric")
plt.xlabel("Log(K) of Number of Components Used")
plt.title("Trade-off between Sparsity and Metrics")

plt.legend()

plt.savefig("bipolar-tradeoff.png")
plt.show()
plt.close()
```
